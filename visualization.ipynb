```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target
```

```python
setosa = data.loc[data['target']==0].reset_index()
versicolor = data.loc[data['target']==1].reset_index()
virginica = data.loc[data['target']==2].reset_index()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(4, 1, figsize=(12, 16))

for idx, feature in enumerate(list(iris.feature_names)):
    axs[idx].plot(setosa[feature], 'r', label='setosa')
    axs[idx].plot(versicolor[feature], 'g', label='versicolor')
    axs[idx].plot(virginica[feature], 'b', label='virginica')
    axs[idx].axhline(y=setosa[feature].mean(), color='r', linestyle='--')
    axs[idx].axhline(y=versicolor[feature].mean(), color='g', linestyle='--')
    axs[idx].axhline(y=virginica[feature].mean(), color='b', linestyle='--')
    axs[idx].grid(True)
    axs[idx].legend()
    axs[idx].set_ylabel(feature)
    axs[idx].set_title(feature)

# Wywietlenie wykresw
plt.show()
```
